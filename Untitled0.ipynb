```python
import tensorflow as tf
from tensorflow import keras
```

Data Definition

```python
# y = 3 * x - 5
x = tf.constant([-2 -1, 0, 1, 2, 5, 7], dtype=float)
y = 3 * x - 5
```

```python
x.numpy()
```

```python
y.numpy()
```

Model Definition

```python
# activation=None --> there is no activation function --> model is linear
model = keras.Sequential([
    keras.layers.Dense(units=1, activation=None, input_shape=[1])
])
```

```python
# optimizer = sgd: stocastic gradient dicend, gd: gradient decend, adam: , rmsprop (try yo optimize weight cost function)
# loss = mean_square_error (it is better the error be derivatiable because optimizer function can easily solve with error)
model.compile(optimizer='sgd', loss='mean_squared_error')
```

```python
model.summary()
```

```python
model.fit(x, y, batch_size=1, epochs=500)
```

```python
# y = 3 * x - 5
x_test = tf.constant([-4, 11, 20])
y_test = x_test * 3 - 5
```

```python
y_test.numpy()
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
import matplotlib.pyplot as plt
plt.scatter(x_test.numpy(), y_test.numpy(), color='red', linewidths=8)
plt.scatter(x_test.numpy(), y_pred, color='blue', linewidths=2)
plt.show()
```
