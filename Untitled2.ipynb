```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
import sklearn
```

# Data Definition

y = x^2 + 5*x + 3

```python
x = np.arange(-80, 81, 0.5)
y = x**2 + 5*x + 3
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
x_train.shape, x_test.shape
```

# Model Definition

```python
model2 = keras.Sequential()
model2.add(keras.layers.Dense(units=256, activation='relu', input_shape=[1]))
model2.add(keras.layers.Dense(units=256, activation='relu'))
model2.add(keras.layers.Dense(units=256, activation='relu'))
model2.add(keras.layers.Dense(units=256, activation='relu'))
model2.add(keras.layers.Dense(units=1, activation=None, input_shape=[1]))
```

```python
optimizer = keras.optimizers.Adam()
# RMSprop(learning_rate=0.0001)
loss = keras.losses.mean_squared_error
metric = tf.metrics.RootMeanSquaredError()
model2.compile(optimizer=optimizer, loss=loss, metrics=metric)
```

```python
# (20% * epochs) <= patience <= (30% * epochs)
early_stop = keras.callbacks.EarlyStopping(monitor='val_loss', patience=300, restore_best_weights=True)
```

```python
hist2 = model2.fit(x_train, y_train, batch_size=64, epochs=1000, callbacks=[early_stop], validation_data=(x_test, y_test))
```

# Comparison

```python
y_pred2 = model2.predict(x_test)
```

```python
plt.scatter(x_test, y_test, color='red', linewidths=6)
plt.scatter(x_test, y_pred2, color='blue')
plt.show()
```

```python
plt.plot(hist2.history['loss'], color='red')
plt.plot(hist2.history['val_loss'], color='blue')
plt.show()
```
