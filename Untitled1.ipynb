```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
import sklearn
```

Data Definition

y = x^2 + 5*x + 3

```python
x = np.arange(-80, 81, 0.5)
y = x**2 + 5*x + 3
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

1. Model Definition

```python
model1 = keras.Sequential()
model1.add(keras.layers.Dense(units=1024, activation='relu', input_shape=[1]))
model1.add(keras.layers.Dense(units=1, activation=None, input_shape=[1]))
```

```python
model1.compile(optimizer='adam', loss='mean_squared_error')
```

```python
model1.summary()
```

```python
hist1 = model1.fit(x_train, y_train, epochs=1000, validation_data=(x_test, y_test))
```

# 2. Model Definition


```python
model2 = keras.Sequential()
model2.add(keras.layers.Dense(units=256, activation='relu', input_shape=[1]))
model2.add(keras.layers.Dense(units=256, activation='relu'))
model2.add(keras.layers.Dense(units=256, activation='relu'))
model2.add(keras.layers.Dense(units=256, activation='relu'))
model2.add(keras.layers.Dense(units=1, activation=None, input_shape=[1]))
```

```python
model2.compile(optimizer='adam', loss='mean_squared_error')
```

```python
model2.summary()
```

```python
hist2 = model2.fit(x_train, y_train, batch_size=64, epochs=1000, validation_data=(x_test, y_test))
```

# Comparison

```python
y_pred1 = model1.predict(x_test)
y_pred2 = model2.predict(x_test)
```

```python
plt.scatter(x_test, y_test, color='red', linewidths=8)
plt.scatter(x_test, y_pred1, color='green')
plt.scatter(x_test, y_pred2, color='blue')
```
